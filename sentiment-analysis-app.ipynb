Chat model to coverse

```python
import os
from dotenv import find_dotenv,load_dotenv
load_dotenv(find_dotenv())

from langchain_openai import ChatOpenAI
llm=ChatOpenAI(
    base_url="https://api.groq.com/openai/v1",
    model="openai/gpt-oss-120b",
    api_key=os.getenv("Free_API_KEY"),
    temperature=0.0,
)
```

## Tag Definition
* In the following code we define the 3 tags we will analize in this app:
    * sentiment.
    * political tendency.
    * language. 

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.pydantic_v1 import BaseModel,Field

# CLass that defines how the llm will response to our input 

class Classification(BaseModel):
    sentiment:str=Field(description="The sentiment of the text")
    political_tendency:str=Field(description="The Political tendency of the user")
    language:str=Field(description="The language the text is written in")
```

```python
# Tagging Prompt 
tagging_prompt=ChatPromptTemplate.from_template(
  """
Extract the desired information from the following passage.

Only extract the properties mentioned in the 'Classification' function.

Passage:
{input}
"""
)
```

```python
model=llm.with_structured_output(
    schema=Classification
)
```

```python
# Creating chain to connect all parts
tagging_chain=tagging_prompt | model
```

```python
response=tagging_chain.invoke({
    "input": "I absolutely love this new phone. The battery life is amazing!"
})
print(response)
```

```python
response = tagging_chain.invoke({
    "input": "This service is terrible and I am very disappointed."
})
print(response)
```

```python
response = tagging_chain.invoke({
    "input": "The government should invest more in education rather than defense."
})
print(response)
```

```python
response = tagging_chain.invoke({
    "input": "मुझे यह फिल्म बहुत पसंद आई।"
})
print(response)
```

* Careful schema definition gives us more control over the model's output.
* Specifically, we can define:
    * Possible values for each property.
    * Description to make sure that the model understands the property.
    * Required properties to be returned.
* Let's redeclare our Pydantic model to control for each of the previously mentioned aspects **using enums**:

```python
class Strict_Classification(BaseModel):
    sentiment:str=Field(description="The sentiment of the text",enum=["happy","neutral","sad"])
    political_tendency:str=Field(description="The Political tendency of the user",enum=["conservative","liberal","independent"])
    language:str=Field(description="The language the text is written in",enum=["Hindi","English"])
```

```python
strict_tagging_prompt=ChatPromptTemplate.from_template(
    """
Extract the desired information from the following passage.

Only extract the properties mentioned in the 'Strict_Classification' function.

Passage:
{input}
"""
)
```

```python
model_b=llm.with_structured_output(
    schema=Strict_Classification
)
```

```python
strict_tagging_chain=strict_tagging_prompt | model_b
```

```python
response=strict_tagging_chain.invoke({
    "input": "I absolutely love this new phone. The battery life is amazing!"
})
print(response)
```

```python
response = strict_tagging_chain.invoke({
    "input": "मुझे यह फिल्म बहुत पसंद आई।"
})
print(response)
```

```python
response = strict_tagging_chain.invoke({
    "input": "This service is terrible and I am very disappointed."
})
print(response)
```
